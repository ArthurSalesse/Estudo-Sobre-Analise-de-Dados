```python
import pandas as pd

# Criando a tabela do RH
dados_rh = {

    "ID": [1, 2, 3],
    "Nome": ["Arthur", "Giovana", "Carlos"]

}

df_rh = pd.DataFrame(dados_rh)

# Criando a tabela do FINANCEIRO

dados_financeiro = {
    "ID": [1, 2, 4],
    "Salario": [5000, 6000, 7000]
}

df_financeiro = pd.DataFrame(dados_financeiro)

df_resultado = pd.merge(df_rh, df_financeiro, on = "ID", how = "left")

# ON = ID, diz qual a chave que liga as tabelas
# HOW = INNER, diz para usar somente informações que existem nas duas tabelas
# HOW = LEFT, mostra todas as informações da tabela mesmo que elas sejam nulas

df_resultado
```

```python
import pandas as pd
import numpy as np # para criar valores NULOS
import matplotlib.pyplot as plt # biblioteca de desenho

# Recebendo os dados sujos
dados_empresa = {

    "Nome": ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"],
    "Salario": [4000, np.nan, 5000, 4500, np.nan], # Bruno e Eduardo não possuem salários
    "Departamento": ["Vendas", "TI", "TI", np.nan, "Vendas"]

}

df = pd.DataFrame(dados_empresa)

df

# Apagando as colunas sem informações
df_limpo_radical = df.dropna()
df_limpo_radical

# Preenchendo os buracos
df_restaurado = df.fillna({
    "Salario": 0,
    "Departamento": "Não informado"
})

resumo_salarial = df_restaurado.groupby("Departamento")["Salario"].sum()

resumo_salarial.plot(kind = 'bar', color = 'skyblue') # tipo de gráfico é barra

# editando o gráfico
plt.title("Total de salários do departamento")
plt.xlabel("Departamento")
plt.ylabel("Valor em Reais")

plt.show()
```

```python
import pandas as pd

# Os dados originais 
dados_vendas = {
    "Mes": ["Jan", "Jan", "Fev", "Fev", "Jan", "Fev"],
    "Regiao": ["Norte", "Sul", "Norte", "Sul", "Norte", "Sul"],
    "Produto": ["Celular", "Celular", "Notebook", "Celular", "Notebook", "Notebook"],
    "Valor": [1000, 1500, 3000, 1200, 2500, 2000]
}
df = pd.DataFrame(dados_vendas)

#  Pivot Table
# values  -> Qual número vamos calcular? (O Valor)
# index   -> O que vai ficar nas linhas? (Ex: Mes)
# columns -> O que vai ficar nas colunas? (Ex: Regiao)
# aggfunc -> O que fazer com os números? ('sum' para somar, 'mean' para média)

tabela_dinamica = pd.pivot_table(df, 
                                 values="Valor", 
                                 index="Mes", 
                                 columns="Regiao", 
                                 aggfunc="sum")

# Mostrar o resultado
tabela_dinamica
```

```python
import pandas as pd

# Dados do sistema
dados_cafe = {

    "Mes": ["Outubro", "Outubro", "Novembro", "Novembro", "Outubro", "Novembro"],
    "Tipo_Cafe": ["Espresso", "Capuccino", "Espresso", "Capuccino","Espresso", "Capuccino"],
    "Quantidade": [50, 30, 80, 120, 40, 60]
}

df = pd.DataFrame(dados_cafe)

tabela_dinamica = pd.pivot_table(df,
                                 values= "Quantidade",
                                 index= "Mes",
                                 columns= "Tipo_Cafe",
                                 aggfunc= "sum")

tabela_dinamica
```

```python
import pandas as pd

# Dados brutos
dados_clientes = {
    "Nome": ["Ana", "Bruno", "Carlos", "Diana"],
    "Anos_Cliente": [6, 2, 7, 1],
    "Total_Gasto": [6000, 1500, 3000, 8000]    
}

df = pd.DataFrame(dados_clientes)

# Criando uma regra
def classificar_clientes(linha):
    if linha["Anos_Cliente"] > 5 and linha["Total_Gasto"] > 5000:
        return "VIP"
    else:
        return "Normal"

# Aplicando a regra
df["Categoria"] = df.apply(classificar_clientes, axis = 1)

df
```

```python
import pandas as pd

# Dados Brutos
dados_dos_clientes = {
    "Nome": ["Alice", "Bruno", "Carla", "Daniel"],
    "Plano_Fidelidade": ["Ouro", "Prata", "Bronze", "Ouro"],
    "Gasto_Total": [1500.00, 600.00, 800.00, 400.00]
}

df_campanha = pd.DataFrame(dados_dos_clientes)

def desconto(linha):
    if linha["Plano_Fidelidade"] == "Ouro" and linha["Gasto_Total"] > 1000.00:
        return 20
    elif linha["Plano_Fidelidade"] == "Prata":
        return 10
    else:
        return 0

df_campanha["Desconto"] = df_campanha.apply(desconto, axis = 1)

df_campanha
```
